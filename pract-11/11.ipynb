```python
import numpy 
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

```python
X=numpy.array([[1],[2],[3],[4],[5],[6],[7],[8]])
y=numpy.array([0,0,0,0,1,1,1,1])
```

```python
model=LogisticRegression()
model.fit(X,y)
```

```python
X_test = numpy.linspace(0,9,100).reshape(-1,1)
y_prob = model.predict_proba(X_test)[:,1]
```

```python
plt.figure(figsize=(8,5))
plt.scatter(X,y,color='black',label="Data Points(Fail=0,Pass=1)")
plt.plot(X_test, y_prob, color="blue" , label="Logistic regression Curve")
plt.axhline(0.5, color='red', linestyle='--', label="Threshold = 0.5")
plt.title("Logistic Regression: Study Hpurs vs. Pass/Fail")
plt.xlabel("Study Hours")
plt.ylabel("Probability of Passing")
plt.legend()
plt.grid(True)
plt.show()
```
